```python
# Рандомная площадь
def rand_p():
    return randrange(20, 500)
```

```python
# случайные даты
def random_date(start, end):

    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60)
    random_second = randrange(int_delta)
    return (start + timedelta(seconds=random_second)).date()


d1 = datetime.strptime('1/1/2008', '%m/%d/%Y')
d2 = datetime.strptime('1/1/2023', '%m/%d/%Y')
```

ДатаФрейм

```python
# количество сущностей (строк)
n = 50000

# рандомные данные для ДФ
data = {
    'Дата': [random_date(d1, d2) for _ in range(n)], 
    'Функциональное назначение': [random.choice(F_OKS) for _ in range(n)], 
    'Информация о видах разрешенного использования земельного участка': [random.choice(ZU) for _ in range(n)], 
    'Субъект РФ': [random.choice(RF) for _ in range(n)],
    'Категории объектов': [random.choice(K_OKS) for _ in range(n)], 
    'Общая площадь объекта': [rand_p() for _ in range(n)]
    } 


# Create DataFrame 
df = pd.DataFrame(data)

# перевод данных столбца "Дата" в формат данных "Дата"
df['Дата'] = pd.to_datetime(df['Дата']).dt.strftime('%d/%m/%Y')
```
